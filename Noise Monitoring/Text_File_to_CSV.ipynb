```python
import pandas as pd
```

```python
file = "/content/drive/MyDrive/FYP/noisedata.txt" # Downloaded text file
newfile = "/content/drive/MyDrive/FYP/noisedata.csv" # CSV file
```

### Obtain list with column names and data only

```python
count = 0
with open (file, "r") as myfile:
  data = myfile.readlines()
```

```python
str_data = str(data)
print(type(str_data))
print(str_data)
```

```python
count = 0
for item in data:
  count += 1
print(count)
```

```python
count = 0
data_column_list = [] # contains all columns names and data values
first_split_data = str_data.split("{") # split data using " {"
for item in first_split_data:
  second_split_data = item.split(",")
  for newitem in second_split_data:
    first_replace = newitem.replace('"',"")
    second_replace = first_replace.replace("}", "")
    third_replace = second_replace.replace("['", "")
    fourth_replace = third_replace.replace("']","")
    third_split = fourth_replace.split(": ")
    for elem in third_split:
      count +=1
      if elem != "": 
        data_column_list.append(elem)
      print(elem)

print(count)
```

```python
print(len(data_column_list))
```

```python
count = 0
column_name_list = [] # contains only column names
data_list = [] # contains only data values
for item in data_column_list:
  if count%2 != 0: # data values are always odd index elements
    data_list.append(item)
    #print(item)
  else:
    column_name_list.append(item) # remaining items are column names
  count += 1

print(count)
```

```python
count = 0

unique_names = []
for x in column_name_list:
  if x not in unique_names:
    unique_names.append(x) # only get unique column names
```

```python
print(unique_names)
```

### Construct Dataframe

```python
num_columns = 0
for item in unique_names:
  num_columns += 1

print(num_columns)
print(len(data_list))
```

```python
# create empty dataframe with only column names
df = pd.DataFrame(columns = [unique_names[0]])
count = 1
for count in range(num_columns):
  column_name = unique_names[count]
  df[column_name] = ""

df
```

```python
count = 0
start_index = 0
column_data = []

for count in range(num_columns):
  index = start_index
  column_name = unique_names[count]
  while index < len(data_list):
    column_data.append(data_list[index])
    index += num_columns
  #print(column_data)
  df[column_name] = column_data
  column_data.clear() # clear the list to append new column of data during next iteration
  count += 1
  start_index += 1

df
```

```python
df.to_csv(newfile,index=False)
```

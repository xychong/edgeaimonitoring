```python
import os
import librosa, librosa.display
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import IPython.display
import scipy.io.wavfile
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
pathname_shout = "/content/drive/MyDrive/FYP Data/Old [Noise]/Shout/Data"

new_pathname_shout = "/content/drive/MyDrive/FYP Data/Old [Noise]/Shout/Trimmed Data"
```

```python
for filename in os.listdir(pathname_shout):
    audiopath = pathname_shout + "/" + filename
    audio_data, sr = librosa.load(audiopath)
    
    plt.figure(figsize = (20,5))
    librosa.display.waveplot(audio_data, sr=sr)
    plt.title(filename)
    plt.xlabel("Time (in seconds)")
    plt.ylabel("Amplitude")
```

```python
for filename in os.listdir(pathname_shout):
    audiopath = pathname_shout + "/" + filename
    audio_data, sr = librosa.load(audiopath)

    trimmed_audio, index = librosa.effects.trim(audio_data, 60)
    
    plt.figure(figsize = (20,5))
    librosa.display.waveplot(trimmed_audio, sr=sr)
    plt.title(filename)
    plt.xlabel("Time (in seconds)")
    plt.ylabel("Amplitude")
```

```python
for filename in os.listdir(pathname_shout):
    audiopath = pathname_shout + "/" + filename
    audio_data, sr = librosa.load(audiopath)

    trimmed_audio, index = librosa.effects.trim(audio_data, 60)

    scipy.io.wavfile.write(new_pathname_shout + "/" + filename, sr, trimmed_audio)
```

```python
for filename in os.listdir(new_pathname_shout):
  audiopath = new_pathname_shout + "/" + filename
  audio, sr = librosa.load(audiopath)
  if librosa.get_duration(audio) !=4 and librosa.get_duration(audio) <3.999:
    print(filename)
    print(librosa.get_duration(audio))
```

```python
# Stretched the audio so that we get a 4s segment
for filename in os.listdir(new_pathname_shout):
  audiopath = new_pathname_shout + "/" + filename
  audio, sr = librosa.load(audiopath)
  if librosa.get_duration(audio) !=4 and librosa.get_duration(audio) <3.999:
    rate = librosa.get_duration(audio)/4.0
    stretched_audio = librosa.effects.time_stretch(audio,rate)
    print(filename)
    print(librosa.get_duration(stretched_audio))
```

```python
# Stretch the audio so that we get a 4s segment
for filename in os.listdir(new_pathname_shout):
  audiopath = new_pathname_shout + "/" + filename
  audio, sr = librosa.load(audiopath)
  if librosa.get_duration(audio) !=4 and librosa.get_duration(audio) <3.999:
    rate = librosa.get_duration(audio)/4.0
    stretched_audio = librosa.effects.time_stretch(audio,rate)
    scipy.io.wavfile.write(pathname_shout + "/" + filename, sr, stretched_audio)
```
